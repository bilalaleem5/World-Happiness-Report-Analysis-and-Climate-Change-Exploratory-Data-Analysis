```python

# Define a function to find outliers

def find_outliers(column):

    Q1 = column.quantile(0.25)

    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR

    upper_bound = Q3 + 1.5 * IQR

    return column[(column < lower_bound) | (column > upper_bound)]


# Iterate through each DataFrame and find outliers

outliers_dict = {}

for year, df in [(2015, df_2015), (2016, df_2016), (2017, df_2017), (2018, df_2018), (2019, df_2019)]:
    year_outliers = {}

    for column in df.columns:

        if df[column].dtype != object: 

            outliers = find_outliers(df[column])

            if not outliers.empty:

                year_outliers[column] = outliers

                    

    if year_outliers:

        outliers_dict[year] = year_outliers     

            
# Print the outliers for each year and column


for year, year_outliers in outliers_dict.items():

    print(f"Outliers in {year}:")

    for column, outliers in year_outliers.items():

        print(f"From file {year}:")

        print(f"{column}:")

        print(outliers)

        
```

```python
# Define a list of data frames and their corresponding year

data_frames = [(df_2015, 2015), (df_2016, 2016), (df_2017, 2017), (df_2018, 2018), (df_2019, 2019)]



# Initialize an empty DataFrame to store the combined data

combined_data = pd.DataFrame()


# for loop condition
for data, year in data_frames:

    # Rename the 'Country or region' column to 'Country' for consistency

    if 'Country or region' in data:
        
        #data storing

        data = data.rename(columns={'Country or region': 'Country'})


    #if condtions
    if 'Happiness.Score' in data:
        #data store
        data = data.rename(columns={'Happiness.Score': 'Happiness Score'})


    #if condition
    if 'Score' in data:
        #DATA STORE
        data = data.rename(columns={'Score': 'Happiness Score'})

    

    # Select only relevant columns

    columns_to_select = ['Country']
    
    #COLUMNS Appending

    columns_to_select.append('Happiness Score')

    
    # Extract and store the data for the current year

    year_data = data[columns_to_select]
    
    #years data

    year_data['Year'] = year  # Add a 'Year' column
    
    #combining data

    combined_data = pd.concat([combined_data, year_data], ignore_index=True)



# Pivot the DataFrame to have years as columns

happiness_df = combined_data.pivot(index='Country', columns='Year', values='Happiness Score').reset_index()



# Rename columns

happiness_df.columns.name = None  


# Remove the column name

happiness_df.columns = ['Country', '2015', '2016', '2017', '2018', '2019']



# Save to CSV

happiness_df.to_csv("Happiness DataFrame.csv", index=False)

#orinting

happiness_df
```
